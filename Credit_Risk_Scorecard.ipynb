```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('loan_data.csv')
```

```python
na_values = df.isna().mean()
na_values[na_values>0.8]
```

```python
columns_to_drop = na_values[na_values > 0.8].index.tolist()
```

```python
df = df.drop(columns_to_drop, axis=1)
```

```python
df.info()
```

```python
df.drop(columns = ['id', 'member_id', 'sub_grade', 'emp_title', 'url', 'title', 'zip_code', 'next_pymnt_d',
                    'recoveries', 'collection_recovery_fee', 'total_rec_prncp', 'total_rec_late_fee'], inplace = True)
```

```python
df.info()
```

```python
df[['emp_length','term',]]
```

```python
#with emp_length and term we will fill any NaN values with 0
#we will then remove everything except the number and change the dtype to numeric
```

```python
# Assuming 'df' is your DataFrame and 'emp_length' is the column to process
df['emp_length'].fillna('0', inplace=True)
df['term'].fillna('0', inplace=True)
df['emp_length'] = df['emp_length'].str.extract('(\d+)')
df['term'] = df['term'].str.extract('(\d+)')
df['emp_length'] = df['emp_length'].astype(int)
df['term'] = df['term'].astype(int)
# Verify the changes
print(df[['emp_length','term']])
```

```python
df['mths_since_last_delinq'] 
```

```python
#we will fill na values with 0 indicating 0 months since delinquency
```

```python
df['mths_since_last_delinq'].fillna(0, inplace=True)
```

```python
print(df['mths_since_last_delinq'])
```

```python
#we do the same for the rest of the int/float variable 
```

```python
df[['revol_util','mths_since_last_major_derog']].fillna(0, inplace=True)
```

```python
df[['issue_d','earliest_cr_line','last_pymnt_d','last_credit_pull_d']]
```

```python
from datetime import datetime
```

```python
df['issue_d'] = pd.to_datetime(df['issue_d'], format='%b-%y')
df['earliest_cr_line'] = pd.to_datetime(df['earliest_cr_line'], format='%b-%y')
df['last_pymnt_d'] = pd.to_datetime(df['last_pymnt_d'], format='%b-%y')
df['last_credit_pull_d'] = pd.to_datetime(df['last_credit_pull_d'], format='%b-%y')

# Check the result
print(df[['issue_d', 'earliest_cr_line', 'last_pymnt_d', 'last_credit_pull_d']].head())
```

```python
date_columns = ['issue_d', 'earliest_cr_line', 'last_pymnt_d', 'last_credit_pull_d']
df[date_columns].info()
```

```python
#We can see that there are null values in two columns. We will replace these with a default date
```

```python
todays_date=pd.to_datetime('2024-08-01')
```

```python
df['last_pymnt_d'].fillna(todays_date,inplace=True)
df['last_credit_pull_d'].fillna(todays_date,inplace=True)
```

```python
df[date_columns].info()
```

```python
#Above we can see that the objects have been converted into date time format and missing values have been imputed
#We will now proceed to convert these columns into 'months since' with the default date as a reference point
```

```python
# Calculate the number of months since each date
for column in date_columns:
    df['months_since_' + column] = ((todays_date - df[column]) / np.timedelta64(1, 'M')).astype(int)
```

```python
# Output the resulting DataFrame
df.head()
```

```python
#Now we can drop the original date columns
```

```python
df.drop(columns=date_columns,inplace=True)
```
